## NMAP

```python
import nmap
```

```python
scanner = nmap.PortScanner()
tcp = "137,138,139,445,3389,21,69,25,143,110,22,23,53,67,68,80,443,161,162,1900,123,5353,514,3306,5432,3050"
udp = "137,138,139,445,67,68,53,69,111,161,162,1900,123,5353,514"
result = scanner.scan("192.168.1.0/24",arguments=f" -sS -sU -p T:{tcp},U:{udp} -T3 -O")
```
